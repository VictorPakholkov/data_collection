```python
import pandas as pd
import numpy as np
from sklearn.metrics import cohen_kappa_score
```

```python
data = pd.read_csv("C:\\Users\\Victo\\Downloads\\data_marked_up.csv")
data.head()
```

```python
data = data.dropna(subset=['classic_1', 'inspiring_1', 'deep_1', 'revolutionary_1', 'nostalgic_1', 'negative_1', 'classic_2', 'inspiring_2', 'deep_2', 'revolutionary_2', 'nostalgic_2', 'negative_2'])
data.head()
```

```python
data[['classic_1', 'inspiring_1', 'deep_1', 'revolutionary_1', 'nostalgic_1', 'negative_1', 'classic_2', 'inspiring_2', 'deep_2', 'revolutionary_2', 'nostalgic_2', 'negative_2']] = data[['classic_1', 'inspiring_1', 'deep_1', 'revolutionary_1', 'nostalgic_1', 'negative_1', 'classic_2', 'inspiring_2', 'deep_2', 'revolutionary_2', 'nostalgic_2', 'negative_2']].applymap(lambda x: x if str(x).replace('.', '').replace('-', '').isdigit() else None)
data = data.dropna(subset=['classic_1', 'inspiring_1', 'deep_1', 'revolutionary_1', 'nostalgic_1', 'negative_1', 'classic_2', 'inspiring_2', 'deep_2', 'revolutionary_2', 'nostalgic_2', 'negative_2'], how='any')
```

```python
data['classic_1'] = data['classic_1'].astype(int)
data['inspiring_1'] = data['inspiring_1'].astype(int)
data['deep_1'] = data['deep_1'].astype(int)
data['revolutionary_1'] = data['revolutionary_1'].astype(int)
data['nostalgic_1'] = data['nostalgic_1'].astype(int)
data['negative_1'] = data['negative_1'].astype(int)
data['classic_2'] = data['classic_2'].astype(int)
data['inspiring_2'] = data['inspiring_2'].astype(int)
data['deep_2'] = data['deep_2'].astype(int)
data['revolutionary_2'] = data['revolutionary_2'].astype(int)
data['nostalgic_2'] = data['nostalgic_2'].astype(int)
data['negative_2'] = data['negative_2'].astype(int)
data.head()
```

```python
labels1 = np.array([[data['classic_1'].values], [data['inspiring_1'].values], [data['deep_1'].values], [data['revolutionary_1'].values], [data['nostalgic_1'].values], [data['negative_1'].values]])  
labels2 = np.array([[data['classic_2'].values], [data['inspiring_2'].values], [data['deep_2'].values], [data['revolutionary_2'].values], [data['nostalgic_2'].values], [data['negative_2'].values]])  

flat_labels1 = labels1.flatten()
flat_labels2 = labels2.flatten()
                   
kappa_value = cohen_kappa_score(flat_labels1, flat_labels2)


print("Cohen's Kappa:", kappa_value)
```
